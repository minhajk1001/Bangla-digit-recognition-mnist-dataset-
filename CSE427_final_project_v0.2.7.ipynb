```python
dataset = [img0, img1, img2, img3, img4, img5, img6, img7, img8, img9]
```

```python
for i in range(10):
    for j in tqdm(range(44358)):
        if dataset[i].iloc[j, 784] == i:
            dataset[i].iloc[j, 784] = 1
        else:
            dataset[i].iloc[j, 784] = 0
dataset
```

```python
TrainTest = {}  
for i in tqdm(range(10)):
    X = (dataset[i].iloc[:,:784]).T
    y = dataset[i].iloc[:,-1]
    y = np.array([y])
    TrainTest['img'+str(i)] = [X,y] 
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
def cost_function(y,m ,hypothesis):
    return 1/m *(-1*(np.sum(y*np.log(hypothesis) + (1-y)*np.log(1-hypothesis))))
```

```python
def gradient_decent(X, y, weights, bias, hypothesis, lamda):
    dw =  1/m * np.dot(hypothesis - y, X.T)
    db =  1/m * np.sum(hypothesis - y)
    weights = weights - lamda * dw
    bias = bias - lamda * db
    
    return weights, bias
```

```python
itter = 1000

#learning rate
lamda = 0.15

#dataset lengths (number of rows)
m = 44358

cost_val = []
trained_parameters = []

for trainsets in range(10):
    X = TrainTest['img' + str(trainsets)][0] 
    y = TrainTest['img' + str(trainsets)][1]
    weights = np.random.randn(1,784)
    bias = 0
    costfun_val = []
    k = 0

    print('Training for dataset '+ str(trainsets))

    for i in tqdm(range(1, itter + 1)):
        z = np.dot(weights, X) + bias 
        hypothesis = sigmoid(z)
        j = cost_function(y,m, hypothesis)
        costfun_val.append(j)
        k += 1
        weights, bias = gradient_decent(X, y, weights, bias, hypothesis, lamda)

    cost_val.append(costfun_val)  
    trained_parameters.append([weights,bias])      
    print('itteration number:', k)
     
```

```python
overall_accuracy = 0
for classes in range(10):
    X = TrainTest['img'+str(classes)][0]
    y = TrainTest['img'+str(classes)][1]

    weights = trained_parameters[classes][0]
    bias = trained_parameters[classes][1]

    true_positive = 0 
    for i in tqdm(range(44358)):
        z = np.dot(weights, X.T.iloc[i,:]) + bias
        hypothesis = sigmoid(z)
        if np.logical_and(hypothesis >= 0.5, y.T[i,0] == 1):
            true_positive += 1
        if np.logical_and(hypothesis < 0.5, y.T[i,0] == 0):
            true_positive += 1    

    #print(true_positive)      
    acc = (true_positive / 44358) * 100
    overall_accuracy = overall_accuracy + acc
    print('accuracy for dataset img'+ str(classes)," = ",acc)
    
print("Model Accuracy: " + str(overall_accuracy/10))
```
